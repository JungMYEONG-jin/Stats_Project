## 당진시 기상 예보 : dj_wth_fc_data

```python
# 당진시 기상 예보
# 전일 20시 기준 예보를 기준으로 작성된 기상 자료
dj_wth_fc_data = dangjin_fcst_data.copy()

# str으로 입력된 시간을 datetime 객체로 변환
dj_wth_fc_data['forecasting_time'] = [dt.datetime.strptime(x[0:13] , '%Y-%m-%d %H') for x in dj_wth_fc_data['Forecast time']]
dj_wth_fc_data['after_time'] = [dt.timedelta(seconds=x*3600) for x in dj_wth_fc_data.forecast]
dj_wth_fc_data['forecasted_time'] = dj_wth_fc_data.forecasting_time + dj_wth_fc_data.after_time

# 20시 기준 예보만 추출
# 당일 20시 기준 예보 중 다음날 자정전까지의 예보만 추출
dj_wth_fc_data = dj_wth_fc_data[dj_wth_fc_data.forecasting_time.dt.hour==20]
dj_wth_fc_data = dj_wth_fc_data[dj_wth_fc_data.forecast <= 25]
dj_wth_fc_data = dj_wth_fc_data.loc[:, ['forecasted_time', 'Temperature', 'Humidity', 'WindDirection', 'Cloud']]
dj_wth_fc_data.columns = ['time', 'Temperature_fc', 'Humidity_fc', 'WindDirection_fc', 'Cloud_fc']
dj_wth_fc_data = dj_wth_fc_data.reset_index(drop=True)

# 기상 예보는 3시간 단위이기에 1시간 단위로 변경
temp_reference = pd.DataFrame()
temp_reference['time'] = pd.date_range(start=dj_wth_fc_data.time[0], end=dj_wth_fc_data.time[len(dj_wth_fc_data)-1], freq='H')
dj_wth_fc_data = pd.merge(temp_reference, dj_wth_fc_data, on='time', how='outer')

# 비어있는 시간의 예보는 선형보간법을 이용
dj_wth_fc_data = dj_wth_fc_data.interpolate()

# 전운량 변수는 범주형으로 변환
dj_wth_fc_data.Cloud_fc = round(dj_wth_fc_data.Cloud_fc)
```

# sequence로 변경

```python
def make_seq(input, sl):
  train_x = []
  train_y = []
  length = len(input)
  for i in range(length-sl):
    train_seq = input[i:i+sl]
    train_label = input[i+sl:i+sl+1]
    train_x.append(train_seq)
    train_y.append(train_label)
  return train_x, train_y
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

from sklearn.preprocessing import MinMaxScaler
```

```python
class simple_lstm(nn.Module):
    
    def __init__(self):
        super().__init__()
        self.input_vector = 1
        self.sequence_length = 24
        self.output_vector = 100
        self.num_layers = 4
        
        self.lstm = nn.LSTM(input_size=self.input_vector, hidden_size=self.output_vector, num_layers=self.num_layers, batch_first=True)
        self.linear = nn.Sequential(
            nn.Linear(self.output_vector, 50),
            nn.Linear(50, 30),
            nn.Linear(30, 10),
            nn.Linear(10,1)
        )
        
    def forward(self, x):
        output, _ = self.lstm(x) #(hidden, cell) 데이터는 사용하지 않음
        return self.linear(output[:,-1,:])
```

```python
input_window =996 #임의의 수
output_window = 24 #168 7일 24시간
window = 12 #window는 12시간 마다는 12시간 마다
num_features = 1 #베이스라인은 feature를 하나만 사용했습니다.
num_power = 1068 # 날짜
end_=168
lstm_units=32
dropout=0.2
EPOCH=30
BATCH_SIZE=128
```

```python
energy.iloc[:,0]
```

```python
train_x=tf.reshape(energy.iloc[:,0].values, [num_power, 24, num_features]) # floating
print(f'train_x.shape:{train_x.shape}')
```

```python
train_window_x= np.zeros(( train_x.shape[0], (train_x.shape[1]-(input_window + output_window))//window, input_window, num_features)) 
train_window_y= np.zeros(( train_x.shape[0], (train_x.shape[1]-(input_window + output_window))//window, output_window, num_features))
print(f'train_window_x.shape:{train_window_x.shape}')
print(f'train_window_y.shape:{train_window_y.shape}')
```

```python
for example in range(train_x.shape[0]):
    
    for start in range(0, train_x.shape[1]-(input_window+output_window), window):
        end=start+input_window
        train_window_x[example, start//window, :] = train_x[example, start: end               , :]
        train_window_y[example, start//window, :] = train_x[example, end  : end+ output_window, :]
```

```python
new_train_x=tf.reshape(train_window_x, [-1, input_window, num_features])
new_train_y=tf.reshape(train_window_y, [-1, output_window,num_features])
print(f'new_train_x.shape:{new_train_x.shape}')
print(f'new_train_y.shape:{new_train_y.shape}')
```

```python
model=Sequential([
LSTM(lstm_units, return_sequences=False),
Dense(output_window * num_features, kernel_initializer=tf.initializers.zeros()), 
Reshape([output_window, num_features])
])
```

```python
model.compile(optimizer='rmsprop', loss='mae', metrics=['mae'])
# 에포크가 끝날 때마다 점(.)을 출력해 훈련 진행 과정을 표시합니다
class PrintDot(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs):
        if epoch % 10 == 0: print('')
        print('.', end='')

#가장 좋은 성능을 낸 val_loss가 적은 model만 남겨 놓았습니다.
save_best_only=tf.keras.callbacks.ModelCheckpoint(filepath="lstm_model.h5", monitor='val_loss', save_best_only=True)


early_stop = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=20)

#검증 손실이 10epoch 동안 좋아지지 않으면 학습률을 0.1 배로 재구성하는 명령어입니다.
reduceLR = tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=10)
```

```python
model.fit(new_train_x, new_train_y, epochs=EPOCH, batch_size=BATCH_SIZE, validation_split = 0.2, verbose=1,
          callbacks=[PrintDot(), early_stop, save_best_only , reduceLR])

model.summary()
```
